```python
from cluster.methods.initialization import mckmeanspp
from cluster.methods import lloyd, compare, sort, hamerly
from cluster.metrics import l2
```

```python
import numpy as np
```

```python
k = 16
d = 2

xs = np.random.normal(size=(10_000, d))
ys = mckmeanspp(xs, k, l2, 100)
```

```python
lb1 = sort(xs, ys, l2, 1000)
```

```python
lb2 = lloyd(xs, ys, l2, 1000)
```

```python
lb3 = hamerly(xs, ys, l2, 1000)
```

```python
import seaborn as sns

sns.scatterplot(x=xs[:, 0], y=xs[:, 1], hue=lb2)
```

```python
np.mean(lb1 == lb2), np.mean(lb1 == lb3)
```
