```python
from cluster import cluster
```

```python
import seaborn as sns
import numpy as np
```

```python
n = 5_000
sqk = 6
k = sqk ** 2

xs = [
    np.random.normal(loc=(i, j), scale=0.2, size=(n // k, 2))
    for i in range(sqk)
    for j in range(sqk)
]

xs = np.vstack(xs)
```

```python
ys, labels = cluster(xs, k=k)
```

```python
sns.scatterplot(x=xs[:, 0], y=xs[:, 1], hue=labels)
```

```python
%timeit cluster(xs, k=k)
```
