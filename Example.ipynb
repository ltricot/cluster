```python
from cluster import cluster
```

```python
import numpy as np
import seaborn as sns
```

```python
xs = [
    np.random.normal(loc=(i, j), scale=0.2, size=(500, 2))
    for i in range(2)
    for j in range(2)
]

xs = np.vstack(xs)
```

```python
ys, labels = cluster(xs, k=4)
```

```python
sns.scatterplot(x=xs[:, 0], y=xs[:, 1], hue=labels)
```

```python
%timeit cluster(xs, k=4)
```
